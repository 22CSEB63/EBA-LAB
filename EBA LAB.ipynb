```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv("diabetes.csv")
features = ['Glucose', 'BMI', 'Insulin', 'Pregnancies']

X = df[features]
Y = df['Outcome']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
model = RandomForestClassifier(n_estimators = 100, random_state = 42)
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
print("Accuracy: ", accuracy_score(Y_test, Y_pred))
print("Classification Report: \n", classification_report(Y_test, Y_pred))

importance = model.feature_importances_
feature_importance = pd.DataFrame({"Feature": features, "Importance": importance})
feature_importance = feature_importance.sort_values(by = "Importance", ascending = False)

plt.figure(figsize = (8, 5))
sns.barplot(data = feature_importance, x = "Importance", y = "Feature")
plt.title("Feature Importance of Predicting Disease")
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder

# Load the data
df = pd.read_csv("heart_disease_uci.csv")  # Replace with your actual file name

# Features and Target
features = ['cp', 'trestbps', 'chol', 'fbs', 'age']


X = df[features].copy()
Y = df['num']
# Encode categorical features
le = LabelEncoder()
X['cp'] = le.fit_transform(X['cp'])      # cp: typical angina, asymptomatic, etc.
X['fbs'] = le.fit_transform(X['fbs'])    # fbs: TRUE/FALSE

    # Handle missing values if any
X = X.fillna(0)

    # Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

    # Model training
model = LogisticRegression(max_iter=100)
model.fit(X_train, Y_train)

    # Prediction and evaluation
Y_pred = model.predict(X_test)
print("✅ Accuracy:", accuracy_score(Y_test, Y_pred))
print("✅ Classification Report:\n", classification_report(Y_test, Y_pred))

    
```

```python
coefficients = pd.DataFrame({"Feature": features, "Coefficient": model.coef_[0]})
coefficients['effect'] = coefficients['Coefficient'].apply(lambda x: "Positive" if x > 0 else "Negative")
coefficients = coefficients.sort_values(by="effect", ascending=False)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=coefficients, x="Coefficient", y="Feature", palette='coolwarm', hue='effect')
plt.title("Impact of Features on Heart Disease")
plt.show()
```

```python
from pulp import LpMaximize, LpProblem, LpVariable
model = LpProblem("Hospital-Bed-Allocation", LpMaximize)
x = LpVariable("GeneralPatients", 0, cat = "Integer")
y = LpVariable("ICUPatients", 0, cat = "Integer")
model += x + y, "TotalPatients"
model += (3 * x) + (7 * y) <= 700, "BedDayConstraint" 
model.solve()
```

```python
print(f"Max Patients Admitted: {model.objective.value()}")
print(f"General Patients Admitted: {x.value()}")
print(f"ICU Patients Admitted: {y.value()}")
```
